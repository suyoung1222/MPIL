```python
image_path = '/content/drive/Shareddrives/2023_ICRA/final_datasets/DFVO/회전/1759_circle_10i/*.png'
save_path = '/content/drive/Shareddrives/2023_ICRA/final_datasets/SFM_learner_3channel/1759_circle/'
```

```python
files = glob.glob(image_path)
```

```python
import cv2
import numpy as np
```

```python
for f in files:
  img = cv2.imread(f)
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  img2 = np.zeros_like(img)
  img2[:,:,0] = gray
  img2[:,:,1] = gray
  img2[:,:,2] = gray

  title, ext = os.path.splitext(f)
  #print(title[80:])
  cv2.imwrite(save_path+title[80:] + '.png', img2)
```
