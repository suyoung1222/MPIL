```python
image_path = '/content/drive/Shareddrives/2023_ICRA/final_datasets/Droid,DSO,ORB/회전/1759_circle/*.png'
save_path = '/content/drive/Shareddrives/2023_ICRA/final_datasets/DFVO/회전/1759_circle_10i/'
```

```python
files = glob.glob(image_path)
```

```python
file_name = []

inputfile = open("/content/drive/Shareddrives/2023_ICRA/final_datasets/Droid,DSO,ORB/회전/timestapme_bumble0223_1759_circle.txt", mode="r", encoding='UTF-8')
lines = inputfile.readlines()
for line in lines:
  tstamps = line.strip().split()
  for tstamp in tstamps:
    tstamp = re.sub("\n", "", tstamp) #remove enter
    file_name.append(tstamp)
inputfile.close()
```

```python
len(file_name)
```

```python
for f in files:
    f_num = int(f[85:104]) # tstamp 파일 명
    img = Image.open(f)
    cnt = 0
    for f_list in file_name:
      if f_num == int(f_list):
        img.save(save_path + "%010i.png" % cnt)
      cnt = cnt + 1


    #img.save(save_path+file_name[f_num - 1]+'.png') # file_name 리스트에는 timestamp명이 순서대로 들어있음
```
